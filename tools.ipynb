```python
from problem import get_train_data, get_test_data
import matplotlib.pyplot as plt
import numpy as np
```

```python
videos_train, labels_train = get_train_data()
```

```python
# Take 30 frames of the video
video= videos_train[0].read_samples(videos_train[0].frame_times[0:299:10])
label=labels_train[0]
```

```python
def plot_frames(num_rows, num_columns, width, height, data):
    
    data = np.reshape(data, (num_rows, num_columns, width, height))
    
    rows_data, columns_data = data.shape[0], data.shape[1]
    
    heights = [frm[0].shape[0] for frm in data]
    widths = [frm.shape[1] for frm in data[0]]

    fig_width = 12.0
    fig_height = fig_width * sum(heights) / sum(widths)
    
    f, axarr = plt.subplots(
        rows_data,
        columns_data,
        figsize=(fig_width, fig_height),
        gridspec_kw={"height_ratios": heights},
    )
    for i in range(rows_data):
        for j in range(columns_data):
            axarr[i, j].imshow(data[i][j], cmap="gray")
            axarr[i, j].axis("off")
    plt.subplots_adjust(wspace=0, hspace=0, left=0, right=1, bottom=0, top=1)
    plt.show()
```

```python
# show the 30 frames in 3 lines of 10 frames
plot_frames(3, 10, 250, 250, video)
print(f"Ground Truth= {label}")
```
